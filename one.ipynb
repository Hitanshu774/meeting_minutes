```python
import os
import requests
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer, BitsAndBytesConfig
import torch
from transformers import pipeline
```

```python
LLAMA = "meta-llama/Llama-3.2-3B-Instruct"
audio_filename = r"C:\Users\hitan\OneDrive\Desktop\sync folder\AI.ML\Projects\meeting_minutes\audio_sample.mp3"
```

```python
#opening the file
audio_file = open(audio_filename, "rb")
```

```python
pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-medium.en",
    dtype=torch.float16,
    # device="cuda",
    return_timestamps=True
)

results = pipe(audio_filename)
transcription = results["text"]
print(transcription)
```

```python
open_source_transcription = transcription
display(Markdown(open_source_transcription))
```

```python
system_message = """
You produce minutes of meetings from transcripts, with summary, key discussion points,
takeaways and action items with owners, in markdown format without code blocks.
"""

user_prompt = f"""
Below is an extract transcript of a Denver council meeting.
Please write minutes in markdown without code blocks, including:
- a summary with attendees, location and date
- discussion points
- takeaways
- action items with owners

Transcription:
{transcription}
"""
messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(LLAMA)
tokenizer.pad_token = tokenizer.eos_token

inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")

streamer = TextStreamer(tokenizer)

model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)

outputs = model.generate(inputs, max_new_tokens=2000, streamer = streamer)
```

```python
response = tokenizer.decode(outputs[0])
```

```python
display(Markdown(response))
```
